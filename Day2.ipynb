```python
def fb_final(limit: int):
    fbs = []
    for i in range(limit):
        fbs.append(str(i))
    fbs[:limit:3] = ["FIZZ"] * len(fbs[:limit:3])
    fbs[:limit:5] = ["BUZZ"] * len(fbs[:limit:5])
    fbs[:limit:15] = ["FIZZBUZZ"] * len(fbs[:limit:15])
    return fbs[1:]
```

```python
fb_final(20)
```

```python
def fizzbuzz1(n:int) -> str:
  if(n%3==0 and n%5==0):
    return "FIZZBUZZ"
  if n%3==0:
    return "FIZZ"
  if n%5==0:
    return "BUZZ"
  return str(n)

def fizzbuzzloop(limit :int) ->[str]:
  return [fizzbuzz1(i) for i in range(1,limit+1)]
```

```python
fizzbuzzloop(120)
```

```python
def is_prime(num):
  for n in range(3,num//2,2):
    if(num%n==0):
      return 'false'
  return 'true'
```

```python
def list_prime(limit:int) ->[int]:
  return [x for x in range(2, limit) if is_prime(x)]

```

```python
list_prime(20)
```

```python
def is_palindrome(n:int)->bool:
  s=str(n)
  return s==s[::-1]

def euler4(start:int,limit:int)->int:
  return [a*b for a in range(start,limit) for b in range(start,limit) if is_palindrome(a*b)]

euler4(100,1000)
```

```python
def rem(n:int)->int:
  return n%10

def is_asc(n:int)->bool:
  if n<10:
    return True
  if rem(n)<=rem(n//10):
    return False
  return is_asc(n//10)
```

```python
is_asc(55554)

def is_asc1(n:int)->bool:
  if n<10:
    return True
  return lambda n%10,n//10:False if n%10<=n//10 else is_asc(n//10)
```

```python
def isasc2(f,n):
  if n<10:
    return True
  if f(n):
    return False
  return isasc2(f,n//10)

isasc2(lambda k:(k%10>=(k//10)%10),678)
  
```
